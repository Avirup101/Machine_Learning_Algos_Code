```python
import numpy as np
x = [1, 2, 5, 4, 7, 2, 8, 13, 17, 21, 24, 25, 12, 14, 13, 17]
```

```python
y = [3, 5, 1,7, 2, 3, 4, 14, 16, 23, 20, 25, 12, 13, 14, 16]
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.show()
```

```python
xmean = np.mean(x)
ymean = np.mean(y)
xmean, ymean
```

```python
x = x - xmean
y = y - ymean
print(x ,y)
plt.scatter(x, y)
```

```python
def reg_line( m,x,y,prev_ssr=999999) :
    alpha = 0.1
    sse = 0
    ypred=[]
    for val in x :
        ypredict = m * val
        ypred.append(ypredict)     
        
    print(ypred)
    
    for i in range(len(y)) :
        sse = sse + ((y[i] - ypred[i]) ** 2)
    #print(sse)
    curr_ssr = sse 
    #print(curr_ssr, prev_ssr)
    if (curr_ssr < prev_ssr) :
        m += alpha
        return reg_line( m,x,y,curr_ssr)
        #return d
    else :
        return ypred

```

```python
#ypred =[]
g=reg_line(0,x,y)
#print("fhaflhadfjahdfjadhfkdfjhadkfjhadfkdf",c)
print("dfsfsdf",g)
plt.scatter(x, y)
plt.plot(x, g)
plt.show()

plt.show()
```
